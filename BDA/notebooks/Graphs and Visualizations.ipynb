```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import os

# Set renderer to browser
pio.renderers.default = 'browser'

# Define a function to process and generate the plot
def plot_sentiment_by_version(file_path):
    # Read the CSV file into DataFrame
    df = pd.read_csv(file_path)
    
    # Mapping sentiment to numeric values
    sentiment_map = {'negative': 1, 'neutral': 3, 'positive': 5}
    df['sentiment_numeric'] = df['sentiment'].map(sentiment_map)
    
    # Drop rows where 'reviewCreatedVersion' or 'sentiment_numeric' is missing
    df = df.dropna(subset=['reviewCreatedVersion', 'sentiment_numeric'])
    
    # Convert 'reviewCreatedVersion' to string for consistency
    df['reviewCreatedVersion'] = df['reviewCreatedVersion'].astype(str)
    
    # Extract the app name from the file path (remove the extension)
    app_name = os.path.splitext(os.path.basename(file_path))[0]
    
    # Group by 'reviewCreatedVersion' and calculate average sentiment
    df_grouped = df.groupby('reviewCreatedVersion').agg({'sentiment_numeric': 'mean'}).reset_index()
    
    # Plot the average sentiment by app version
    fig = px.line(
        df_grouped,
        x='reviewCreatedVersion',
        y='sentiment_numeric',
        title=f'Average Sentiment by App Version ({app_name})',  # Use app name here
        labels={'reviewCreatedVersion': 'App Version', 'sentiment_numeric': 'Average Sentiment'}
    )
    return fig

# Create a list to store all the plots
all_figures = []

# List of dataset file paths (now with 4 datasets)
dataset_paths = [
    "C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/Insta_output_sentiment.csv",
    "C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/X_output_sentiment.csv",
    "C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/Reddit_output_sentiment.csv",
    "C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/snapchat_output_sentiment.csv"
]

# Loop through each dataset and generate the plots
for file_path in dataset_paths:
    fig = plot_sentiment_by_version(file_path)
    all_figures.append(fig)

# Combine all the figures into one HTML file
all_html = ""
for fig in all_figures:
    all_html += fig.to_html(full_html=False, include_plotlyjs='cdn')

# Save all graphs as one HTML file
with open("all_version_sentiment_graphs.html", "w") as f:
    f.write(all_html)

# The combined HTML file can be opened in the browser
print("All graphs saved in 'all_version_sentiment_graphs.html'. Open it to view all graphs.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
from datetime import datetime

# Load the dataset
df = pd.read_csv("C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/Insta_output_sentiment.csv")

# Ensure `at` column is in datetime format and handle missing content
df['at'] = pd.to_datetime(df['at'], errors='coerce')
df['content'] = df['content'].fillna('')

# Initialize Dash app
app = dash.Dash(__name__)

# App layout
app.layout = html.Div([
    html.H1("Keyword and Topic Analysis Dashboard", style={'textAlign': 'center'}),
    
    # Filters
    html.Div([
        html.Label("Select Sentiment:"),
        dcc.Dropdown(
            id='sentiment-dropdown',
            options=[
                {'label': 'Positive', 'value': 'positive'},
                {'label': 'Neutral', 'value': 'neutral'},
                {'label': 'Negative', 'value': 'negative'}
            ],
            value='positive',  # Default to Positive Sentiment
            clearable=False
        ),
        
        html.Label("Select Date Range:"),
        dcc.DatePickerRange(
            id='date-picker',
            start_date=df['at'].min(),
            end_date=df['at'].max(),
            display_format='YYYY-MM-DD'
        )
    ], style={'marginBottom': '20px'}),
    
    # Word Cloud and Top Keywords
    html.Div([
        dcc.Graph(id='word-cloud', style={'display': 'inline-block', 'width': '49%'}),
        dcc.Graph(id='top-keywords', style={'display': 'inline-block', 'width': '49%'})
    ])
])

# Callback for interactivity
@app.callback(
    [Output('word-cloud', 'figure'),
     Output('top-keywords', 'figure')],
    [Input('sentiment-dropdown', 'value'),
     Input('date-picker', 'start_date'),
     Input('date-picker', 'end_date')]
)
def update_graphs(selected_sentiment, start_date, end_date):
    # Filter data by sentiment and date range
    filtered_df = df[
        (df['sentiment'] == selected_sentiment) &
        (df['at'] >= pd.to_datetime(start_date)) &
        (df['at'] <= pd.to_datetime(end_date))
    ]
    
    # Extract text content
    text_data = " ".join(filtered_df['content'])
    
    # Generate Word Cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)
    wordcloud_fig = px.imshow(wordcloud, template='simple_white', title=f"Word Cloud for {selected_sentiment.capitalize()} Sentiment")
    wordcloud_fig.update_layout(coloraxis_showscale=False)
    wordcloud_fig.update_xaxes(visible=False)
    wordcloud_fig.update_yaxes(visible=False)
    
    # TF-IDF for Top Keywords
    tfidf = TfidfVectorizer(max_features=10, stop_words='english')  # Limit to 10 top keywords
    tfidf_matrix = tfidf.fit_transform(filtered_df['content'])
    tfidf_scores = pd.DataFrame({
        'keyword': tfidf.get_feature_names_out(),
        'score': tfidf_matrix.sum(axis=0).tolist()[0]
    }).sort_values(by='score', ascending=False)
    
    # Bar chart for top keywords
    keywords_fig = px.bar(
        tfidf_scores,
        x='keyword',
        y='score',
        text='score',
        title=f"Top Keywords for {selected_sentiment.capitalize()} Sentiment",
        labels={'score': 'TF-IDF Score', 'keyword': 'Keyword'}
    )
    keywords_fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
    
    return wordcloud_fig, keywords_fig

# Run app
if __name__ == '__main__':
    app.run_server(debug=True, port=8057)  # Change to another port number
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Load the CSV files into separate DataFrames
df1 = pd.read_csv("C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/Insta_output_sentiment.csv")
df2 = pd.read_csv("C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/X_output_sentiment.csv")
df3 = pd.read_csv("C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/Reddit_output_sentiment.csv")
df4 = pd.read_csv("C:/Users/smart/Downloads/Code-20241223T124045Z-001/Data/Sentiment analysis data/Sentiment analysis data/snapchat_output_sentiment.csv")  # Load Snapchat data

# Define the bins for the scores 1, 2, 3, 4, 5
bins = [1, 2, 3, 4, 5, 6]

# Create a figure with 4 subplots
fig, axes = plt.subplots(1, 4, figsize=(20, 5), sharey=True)

# Plot the histogram for Instagram
sns.histplot(df1['score'], bins=bins, kde=False, discrete=True, ax=axes[0])
axes[0].set_title('Instagram Distribution')
axes[0].set_xlabel('Score')
axes[0].set_xticks([1, 2, 3, 4, 5])

# Plot the histogram for X
sns.histplot(df2['score'], bins=bins, kde=False, discrete=True, ax=axes[1])
axes[1].set_title('X Distribution')
axes[1].set_xlabel('Score')
axes[1].set_xticks([1, 2, 3, 4, 5])

# Plot the histogram for Reddit
sns.histplot(df3['score'], bins=bins, kde=False, discrete=True, ax=axes[2])
axes[2].set_title('Reddit Distribution')
axes[2].set_xlabel('Score')
axes[2].set_xticks([1, 2, 3, 4, 5])

# Plot the histogram for Snapchat
sns.histplot(df4['score'], bins=bins, kde=False, discrete=True, ax=axes[3])
axes[3].set_title('Snapchat Distribution')
axes[3].set_xlabel('Score')
axes[3].set_xticks([1, 2, 3, 4, 5])

# Set shared ylabel
axes[0].set_ylabel('Number of Reviews')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a figure with 4 subplots
fig, axes = plt.subplots(1, 4, figsize=(24, 6), sharex=True)

# Plot the countplot for Instagram
sns.countplot(y='reviewCreatedVersion',
              data=df1,
              order=df1['reviewCreatedVersion'].value_counts().index[:10],
              ax=axes[0])
axes[0].set_title('Instagram (Top 10 Versions)')
axes[0].set_xlabel('Number of Reviews')
axes[0].set_ylabel('App Version')

# Plot the countplot for X
sns.countplot(y='reviewCreatedVersion',
              data=df2,
              order=df2['reviewCreatedVersion'].value_counts().index[:10],
              ax=axes[1])
axes[1].set_title('X (Top 10 Versions)')
axes[1].set_xlabel('Number of Reviews')
axes[1].set_ylabel('')

# Plot the countplot for Reddit
sns.countplot(y='reviewCreatedVersion',
              data=df3,
              order=df3['reviewCreatedVersion'].value_counts().index[:10],
              ax=axes[2])
axes[2].set_title('Reddit (Top 10 Versions)')
axes[2].set_xlabel('Number of Reviews')
axes[2].set_ylabel('')

# Plot the countplot for Snapchat
sns.countplot(y='reviewCreatedVersion',
              data=df4,
              order=df4['reviewCreatedVersion'].value_counts().index[:10],
              ax=axes[3])
axes[3].set_title('Snapchat (Top 10 Versions)')
axes[3].set_xlabel('Number of Reviews')
axes[3].set_ylabel('')

# Adjust layout
plt.tight_layout()
plt.show()
```
